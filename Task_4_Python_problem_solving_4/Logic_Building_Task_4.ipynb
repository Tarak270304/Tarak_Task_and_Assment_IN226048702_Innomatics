Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 199:
        print("Minimum recharge plan is ₹199.")
        return False

    if amount not in valid_plans:
        print("Invalid plan. Available plans:", valid_plans)
        return False

    return True


while True:
    recharge_amount = int(input("Enter the recharge amount: ₹"))

    if validate_recharge(recharge_amount):
        print("Recharge successful for ₹", recharge_amount)
        break   # this stops the loop
    else:
        print("Please try again.\n")
```

Problem 2: Inventory Reorder Alert System

```python
#inventory
def check_inventory(stock):
    for product,quantity in stock.items():
        if quantity <15:
            print(f"Reorder alert:{product} (stock:{quantity})")
        else:
            print("stock ok:{product} (stock:{quantity})")
inventory={
    "Rice":10,
    "Wheat":15,
    "Sugar":20,
    "oil":20,
    "salt":3
    
}
check_inventory(inventory)
```

 Problem 3: Student Result Processing System

```python
def check_result(marks):
    total=0
    for mark in marks:
        total+=mark
    average=total/len(marks)
    if average>50:
        return f"pass (average:{average})"
    else:
        return f"fail (average:{average})"


marks=[55,90,65,99,25]

result=check_result(marks)
print(result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if is_peak:
        fare += fare * 0.25   # 25% extra

    return fare


while True:
    distance = float(input("Enter distance in km: "))

    peak_input = input("Is it peak hour? (yes/no): ").lower()
    is_peak = True if peak_input == "yes" else False

    total_fare = calculate_fare(distance, is_peak)

    print(f"Total Cab Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance):
    present = 0

    for i in attendance:
        if i == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance_list)
print(result)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit=False
    has_special=False
    special_chars = "@#$"

    if len(password)<8:
        return "weak password"

    for ch in password:
        if ch.isdigit():
            has_digit=True
        if ch in special_chars:
            has_special=True

        if has_digit and has_special:
            return "Strong Password"
        else:
            return "Weak Password"

pwd=input("Enter the Password:")
result=check_password(pwd)
print(result)
```
