## Task 1: User Login Check
Checks whether the entered username and password are valid.


```python
# User input
username=input("Enter the username:");
password=input("enter password:");

# Stored credentials
stored_user="admin";
stored_password="1234";

# Login check
if username == stored_user and password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

Task 2: Pass / Fail Analyzer

```python
# Student marks
marks = [45, 78, 90, 33, 60]

passed_count = 0
failed_count = 0

# Analyze results
for i in marks:
    if i >= 50:
        passed_count+=1
    else:
        failed_count+=1


# Final result
print("total students passed = ",passed_count)
print("total students failed = ",failed_count)
```

Task 4: Message Length Analyzer

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze message lengths
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("→ Flag: Message longer than 10 characters\n")
    else:
        print("→ Status: Message length is acceptable\n")
```

Task 5: Error Message Detector

```python
# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
    if i == "ERROR":
        count+=1
print("Number of ERROR entries:",count)


```
