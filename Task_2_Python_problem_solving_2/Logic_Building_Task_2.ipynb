Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
count=0
total_days=len(attendance)
for i in attendance:
    if i == "P":
        count+=1
Attendance_Percentage = (count/total_days)*100
print("Attendance Percentage:",Attendance_Percentage)
```

Problem 6: Remove Duplicate Phone Numbers

```python
# Take input from user
phone_numbers = input("Enter phone numbers separated by space: ")
# Convert input string to list of integers
phone_list=list(map(int,phone_numbers.split()))
# Remove duplicates using set
unique_numbers=set(phone_list)
# Display result
print("Unique phone numbers:", unique_numbers)
```

Problem 7: Count Character Frequency
